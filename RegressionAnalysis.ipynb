```python
#### Zahtjeva instalaciju xlrd paketa.

import pandas
import matplotlib.pyplot as plt
from scipy import stats

#### Dohvatanje podataka it Excel fajla.
#### Izvor:
######  https://stackoverflow.com/questions/28774960/how-to-get-read-excel-data-into-an-array-with-python
data = pandas.read_excel('bottle.xlsx')

#### Izbacujemo NA vrijednosti (prazne ćelije).
#### Izvor:
######  https://stackoverflow.com/questions/22551403/python-pandas-filtering-out-nan-from-a-data-selection-of-a-column-of-strings
data.dropna(inplace=True)

#### Izvlačimo podatke i provjeravamo da li trebamo da ih transponujemo (ako su u kolonama).

data = data.values
if data.size > data[0].size:
    data = data.transpose()

#### Razdvajamo redove sa podacima.

x = data[0]
y = data[1]

#### Izvor:
######  https://towardsdatascience.com/simple-and-multiple-linear-regression-in-python-c928425168f9
#### Objašnjenja:
######  y = kx + n = slope * x + intercept
######  p_value, r_value and std_err su parametri koji se mogu koristiti za dalju analizu.

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

#### Ispisujemo originalne podatke i linearnu finkciju unutar našeg opsega.
###### Koristimo zorder parametar da gurnemo podatke prema pozadini, da bismo vidjeli naše podatke i njihove procjene greške.
plt.plot(x, y, 'o', label='original data', zorder=-32)
plt.plot(x, intercept + slope * x, 'r', label='fitted line')

#### Ispisujemo tri primjera podataka: pri sredini opsega, na početku opsega i nakon kraja (van) opsega.
#### Izvor:
######  https://matplotlib.org/1.2.1/examples/pylab_examples/errorbar_demo.html

y_err_base = y.max() - y.min()
r_squared = r_value * r_value

x_mid = x.mean()
y_mid = intercept + slope * x_mid

x_min = x.min()
y_min = intercept + slope * x_min

x_out = x.max() + x_mid * 0.1
y_out = intercept + slope * x_out

plt.errorbar(x_mid, y_mid, xerr=None, yerr=y_err_base * r_squared, fmt='x', ecolor='b', capthick=2, capsize=10, label='mid err')
plt.errorbar(x_min, y_min, xerr=None, yerr=y_err_base * r_squared, fmt='x', ecolor='g', capthick=2, capsize=10, label='min err')
plt.errorbar(x_out, y_out, xerr=None, yerr=y_err_base * r_squared, fmt='x', ecolor='r', capthick=2, capsize=10, label='out err')

plt.legend()
plt.show()
```
